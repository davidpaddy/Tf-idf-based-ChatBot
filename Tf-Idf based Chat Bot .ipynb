```python
#!pip install datasets
#We are gonna make use of one of hugging face's datasets
from datasets import load_dataset

dataset = load_dataset("tatsu-lab/alpaca")
```

```python
#Appending everything into a list called data
data=[]
for a,i,j in zip(dataset['train']['input'],dataset['train']['instruction'],dataset['train']['output']):
    data.append(i + a)
    data.append(j)
```

```python
#output for data list
data
```

```python
questions = data[::2]
```

```python
#example of question list
questions[436]
```

```python
answers = data[1::2]
```

```python
#example of answer list
answers[436]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(questions)
def retrieve_answer(user_input, tfidf_matrix, answers):
    user_input_tfidf = vectorizer.transform([user_input])
    similarities = cosine_similarity(user_input_tfidf, tfidf_matrix)[0]
    most_similar_idx = similarities.argmax()
    return answers[most_similar_idx]
while True:
    user_input = input("Question: ")
    if user_input.lower() == "exit":
        break
    response = retrieve_answer(user_input, tfidf_matrix, answers)
    print("Chatbot:", response)
```
